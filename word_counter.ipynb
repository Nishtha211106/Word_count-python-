```python
sentence=input("Enter a sentence:")
```

```python
splited=sentence.split()
```

```python
my_dict={}
```

```python
for word in splited:
    if word in my_dict:
        my_dict[word]+=1
    else:
        my_dict[word]=1
```

```python
total_items=len(splited)
unique_items=len(my_dict)
```

```python
max_value=max(my_dict.values())
#max_key = max(data, key=data.get) to get the key corresponding to max value
#max_key = max(my_dict, key=my_dict.get)
```

```python
for word, count in my_dict.items():
  if(count==max_value):
    print(word + ": " + str(count))
```

```python
print("Total number of items: "+str(total_items))
print("Total number of unique items: "+str(unique_items))
print("The words which appeared max number of times is: ")
for word, count in my_dict.items():
  if(count==max_value):
    print(word + ": " + str(count))
print("They appreared: "+str(max_value)+" times")
print("Words with their count in dictionary")
for word, count in my_dict.items():
  print(word + ": " + str(count))
```
